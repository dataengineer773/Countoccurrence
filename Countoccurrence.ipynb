```python
needles = ['apple', 'banana', 'cherry']  # Define the list of items you're searching for
haystack = ['banana', 'orange', 'apple', 'grape']  # Define the list you're searching in

found = 0
for n in needles:
    if n in haystack:
        found += 1

print(f"Found {found} needles in the haystack.")
```
